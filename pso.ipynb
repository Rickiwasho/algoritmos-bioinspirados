# Particle Swarm Optimization


El enjambre consiste de N partículas volando sobre un espacio de dos dimensiones. Cada partícula i es un posible candidato a la solución del problema y está representado por el vector $x_i$ en el espacio de búsqueda. 

function: 

$f(x) = 10n + \sum \limits _{i=1} ^{n} [x^2_i - 10cos(2\pi x_i)]$


```python
import math
import numpy as np
import random as rand
import matplotlib.pyplot as plt

def rastrigin(*X):
        return 10*2 + sum([(x**2 - 10 * np.cos(2 * math.pi * x)) for x in X])

W = 0.5
c1 = 0.8
c2 = 0.9
n_iterations = 50
n_particles = 30
target_error = 1e-6
```

```python
class Particle():
    def __init__(self):
        x = (-1)**(bool(rand.getrandbits(1))) * rand.random()*1000
        y = (-1)**(bool(rand.getrandbits(1))) * rand.random()*1000

        self.position = np.array([x,y])
        self.pbest_position = self.position
        self.pbest_value = float('inf')
        self.velocity = np.array([0,0])
    
    def update(self):
        self.position = self.position + self.velocity
        
```

```python
class Space():
    def __init__(self, target, target_error, n_particles):
        self.target = target
        self.target_error = target_error
        self.n_particles = n_particles
        self.particles = []
        self.gbest_value = float('inf')
        self.gbest_position = np.array([rand.random()*50 , rand.random()*50])
        
    def fitness(self, particle):
        x = particle.position[0]
        y = particle.position[1]
        f = rastrigin(2)
        
        #f = x**2 + y**2 + 1 
        
        return f
    
    
    
    def set_pbest(self):
        for particle in self.particles:
            best_fitness_candidate = self.fitness(particle)
            if(self.gbest_value > best_fitness_candidate):
                self.gbest_value = best_fitness_candidate
                self.gbest_position = particle.position
    
    def set_gbest(self):
        for particle in self.particles:
            best_fitness_candidate = self.fitness(particle)
            if(self.gbest_value > best_fitness_candidate):
                self.gbest_value = best_fitness_candidate
                self.gbest_position = particle.position
                
    def update_particles(self):
        for particle in self.particles:
            global W
            inertial = W + particle.velocity
            self_confidence = c1 * rand.random() * (particle.pbest_position - particle.position)
            swarm_confidence = c2 * rand.random() * (self.gbest_position - particle.position)
            new_velocity = inertial + self_confidence + swarm_confidence
            particle.velocity = new_velocity
            particle.update()
    
    def show_particles(self, iteration):
        print(iteration, " iterations")
        print("best position in this time: ", self.gbest_position)
        print("best value in this time: ", self.gbest_value)
        for particle in self.particles:
            plt.plot(particle.position[0], particle.position[1], "ro")
        plt.plot(self.gbest_position[0], self.gbest_position[1], "bo")
        plt.show()
```

```python
search_space = Space(1, target_error, n_particles)
particle_vector = [Particle() for _ in range(search_space.n_particles)]
search_space.particles = particle_vector

iteration = 0
while(iteration < n_iterations):
    #set particle best & global best
    search_space.set_pbest()
    search_space.set_gbest()
    
    #visualization
    search_space.show_particles(iteration)
    
    #check conditional
    if(abs(search_space.gbest_value - search_space.target) <= search_space.target_error):
        break

    search_space.update_particles()
    iteration += 1
    
    print("the best solution is: ", search_space.gbest_position, " in ", iteration, " iterations. ")
```
